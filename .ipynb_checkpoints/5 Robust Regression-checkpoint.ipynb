## Robust Regression:

Robust regression methods, such as Huber regression, minimize a combination of squared and absolute differences. They are designed to be less influenced by outliers.

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import HuberRegressor, LinearRegression
from sklearn.metrics import mean_squared_error

# Generate some example data with outliers
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + 1.5 * np.random.randn(100, 1)
```

```python
# Introduce outliers
y[70:] += 20 * np.random.randn(30, 1)
```

```python
# Fit a linear regression model (for comparison)
lin_reg = LinearRegression()
lin_reg.fit(X, y)

# Fit a Huber regression model
huber_reg = HuberRegressor(epsilon=1.35)  # epsilon is a tuning parameter
huber_reg.fit(X, y)

# Make predictions on the test set
y_pred_lin = lin_reg.predict(X)
y_pred_huber = huber_reg.predict(X)

# Calculate and print the mean squared error for both models
mse_lin = mean_squared_error(y, y_pred_lin)
mse_huber = mean_squared_error(y, y_pred_huber)
print(f"Linear Regression MSE: {mse_lin}")
print(f"Huber Regression MSE: {mse_huber}")

# Plot the original data points
plt.scatter(X, y, color='red', marker='o', label='Original Data')

# Plot the linear regression line
plt.plot(X, y_pred_lin, label='Linear Regression', linewidth=2)

# Plot the Huber regression line
plt.plot(X, y_pred_huber, label='Huber Regression', linewidth=2)

plt.xlabel('X')
plt.ylabel('y')
plt.title('Robust Regression Example')
plt.legend()
plt.show()
```
